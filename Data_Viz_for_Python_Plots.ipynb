```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('figure', figsize=(20.0, 10.0))
```

```python
df = pd.read_csv('Shark Tank Companies.csv')
```

```python
df.describe()
```

```python
plt.hist(df['valuation'])
xtick_val = [0,10000000, 20000000, 30000000]
xtick_lab = ['0MM','10M', '20M', '30M'] 
plt.xticks(xtick_val, xtick_lab)
plt.xlabel('Valuation')
plt.ylabel('Frequency (Hundreds)')
```

```python
plt.suptitle('Frequent Company Categories landing deals')
df.category.value_counts().plot(kind = 'bar')
plt.show()
subset_deals = df[df['deal'] == True]
count_cat = subset_deals['category'].value_counts()
count_cat.head(10).plot(kind = 'bar')
subset_no_deals = df[df['deal'] == False]
count_cat_no_deals = subset_no_deals['category'].value_counts()
plt.title('Categories That Land Deals Frequently')
plt.show()
count_cat_no_deals.head(10).plot(kind = 'bar')
plt.title('Categories That Do Not Land Deals Frequently')
```

```python
df.boxplot(column = 'valuation', by = 'category')
ytick_val = [0,10000000, 20000000, 30000000]
ytick_lab = ['0MM','10M', '20M', '30M'] 
plt.yticks(ytick_val, ytick_lab)
plt.xticks(rotation=90)
```

```python
plt.rc('figure', figsize=(20.0, 10.0))
plt.subplot(1, 3, 1)
a = df.loc[(df.category == 'Specialty Food')]
b = df.loc[(df.category == 'Novelties')] 
c = df.loc[(df.category == 'Baby and Child Care')]
plt.suptitle('Valuation of Top 3 Categories')
plt.hist(a['valuation'])
plt.xlabel('Speciality Food')
plt.show()
plt.subplot(1,3,2)
plt.hist(b['valuation'])
plt.xlabel('Novelties')
plt.show()
plt.subplot(1,3,3)
plt.hist(c['valuation'])
plt.xlabel('Baby and Child Care')
plt.show()
```
